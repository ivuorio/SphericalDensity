```python
import numpy as np
from scipy.io import FortranFile
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d
from itertools import product
```

```python
#list for sperical harmonics coefficients
chdl  = []
chd0  = []  
ri    = []

#read the unformatted FORTRAN file
f = FortranFile('chdlsoft.DNS','r')

#print all the "line" in the file
stop=True
i=1
while(stop):

    try:
        chdl.append(f.read_reals(dtype=np.float64))
    except:
        stop = False
    #print('read', i,'th set of chdl coefs')
    i+=1
f.close()

#read the unformatted FORTRAN file
f = FortranFile('r.DNS','r')

#print all the "line" in the file
stop=True
i=1
while(stop):

    try:
        ri.append(f.read_reals(dtype=np.float64))
    except:
        stop = False
    #print('read', i,'th set of ri')
    i+=1
f.close()

#read the unformatted FORTRAN file
f = FortranFile('chd0soft.DNS','r')

#print all the "line" in the file
stop=True
i=1
while(stop):

    try:
        chd0.append(f.read_reals(dtype=np.float64))
    except:
        stop = False
    #print('read', i,'th set of chd0 coefs')
    i+=1
f.close()
```

```python
Lms = []
for l in range(9):
    for m in range(-l,l+1):
        lm = l**2+l+m+1
        #print('l, m, L = {0:3} {1:3} {2:7}'.format(l, m, lm))
        Lms.append([l ,m])
#for lm in range(49):
    #print(Lms[lm])
```

```python
from scipy.special import sph_harm

def real_harm(l,m,theta, phi):
    if m == 0:
        return np.real(sph_harm(m,l,phi,theta))
    elif m < 0:
        m = abs(m)
        return (-1)**m * np.sqrt(2) * np.imag(sph_harm(m,l,phi,theta))
    else:
        return (-1)**m * np.sqrt(2) * np.real(sph_harm(m,l,phi,theta))
```

```python
nq = 1 #ATOMIC SITE
#neglect the CPA's atomic site index number
nta = 1
ns = 1 #SPIN INDEX
nlmf = 49#number of spherical harmonics used in the expansion

RSteps,ThetaSteps,PhiSteps = 364,100,100
Rs = range(RSteps) #number of mesh point
Phis = np.linspace(0,np.pi,PhiSteps)
Thetas = np.linspace(0,2*np.pi,ThetaSteps)
RMesh,PhiMesh,ThetaMesh = np.meshgrid(Rs,Phis,Thetas)
#print(PhiMesh)
density = []
z3DFull = np.zeros((ThetaSteps,PhiSteps,RSteps))
#print (ri)

#Loop over atomic sites
for iq in range(1,nq+1):
    #loop atomic site index
    for ita in range(1,nta+1):
        #loop spin index
        for iS in range(1,ns+1):
            density1 = chdl[0][RMesh]*np.abs(real_harm(Lms[0][0],Lms[0][1],PhiMesh,ThetaMesh))
            density = density1.copy()
            #loop the sperical harmonics
            for lm in range(2,nlmf+1):
                #print (RMesh[1],PhiMesh[1],ThetaMesh[1])
                #print(chdl[lm-1][RMesh]*np.abs(real_harm(Lms[lm-1][0],Lms[lm-1][1],PhiMesh,ThetaMesh)))
                density += chdl[lm-1][RMesh]*np.abs(real_harm(Lms[lm-1][0],Lms[lm-1][1],PhiMesh,ThetaMesh))
                print ('done: ',lm)
                """
                for t in range(ThetaSteps):
                    theta=ThetaMesh[t]
                    #
                    for p in range(PhiSteps):
                        phi=PhiMesh[p]
                        for r in RMesh:
                            R = ri[0][r]
                            #print (t,p,r)
                            #print (theta,phi,R)
                            z3DFull[t][p][r] += chdl[lm-1][(r)]*np.abs(real_harm(Lms[lm-1][0],Lms[lm-1][1],phi,theta))
                            #print(z3DFull[t][p][r])
                """
np.save('density',density)
print (density) #muotoa [[[etäisyys]phi]theta]
```
